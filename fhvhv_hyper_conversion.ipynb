```python
hyper_file = "/content/drive/MyDrive/yellow_trip_data/combined_yellow.hyper"
csv_file = "/content/drive/MyDrive/yellow_trip_data/combined_yellow.csv"  # Fixed path

from tableauhyperapi import HyperProcess, Connection, Telemetry
import csv

try:
    with HyperProcess(telemetry=Telemetry.SEND_USAGE_DATA_TO_TABLEAU) as hyper:
        with Connection(hyper.endpoint, hyper_file) as connection:
            # Check available schemas (optional)
            schemas = connection.catalog.get_schema_names()
            print(f"Available schemas: {schemas}")

            schema_name = "Extract"  # or "public" if Extract doesn't work
            tables = connection.catalog.get_table_names(schema_name)

            if not tables:
                raise ValueError(f"No tables found in schema '{schema_name}'.")

            table_name = tables[0]
            print(f"Exporting table: {table_name}")

            # Get column definitions
            table_def = connection.catalog.get_table_definition(table_name)

            # Export to CSV
            with connection.execute_query(f'SELECT * FROM {table_name}') as result:
                with open(csv_file, 'w', newline='', encoding='utf-8') as csvfile:
                    writer = csv.writer(csvfile)
                    writer.writerow([col.name for col in table_def.columns])  # Header
                    writer.writerows(result)  # Data

    print(f"✅ Successfully exported to {csv_file}")

except Exception as e:
    print(f"❌ Error: {e}")
```
